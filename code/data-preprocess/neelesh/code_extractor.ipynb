```python
import pandas as pd
import os
import requests as rq
from bs4 import BeautifulSoup
import pickle
```

```python
DATASET_PATH = '../../../data/stack-overflow'
DATASET_FILE_NAME = 'Dataset - Pandas.csv'
BODY_OUTPUT_FOLDER = 'answer_bodies'
CODE_OUTPUT_FOLDER = 'answer_codes'
SO_API_FIRST_PART = 'https://api.stackexchange.com/2.2/answers/'
SO_API_SECOND_PART = '?order=desc&sort=activity&site=stackoverflow&filter=withbody'
```

```python
def save_object(obj, filename):
    with open(filename, 'wb') as output:  # Overwrites any existing file.
        pickle.dump(obj, output, pickle.HIGHEST_PROTOCOL)
```

```python
df = pd.read_csv(os.path.join(DATASET_PATH, DATASET_FILE_NAME))
save_object(df, 'dataset_df.pkl')

for idx, row in df.iterrows():
    print('Processing row --> '+str(idx))
    answer_id = row['AnswerId']
    try:
        if answer_id != 0:
            # Download the answer post identified by answer id using StackExchange API
            resp = rq.get(SO_API_FIRST_PART+str(answer_id)+SO_API_SECOND_PART)
        
            # Write the bodies to the local files
            with open(os.path.join(BODY_OUTPUT_FOLDER, str(answer_id)), 'w') as body_text:
                body_text.write(resp.json()['items'][0]['body'])
    except:
        print answer_id
```

```python
for fil in os.listdir(BODY_OUTPUT_FOLDER):
    f = open(os.path.join(BODY_OUTPUT_FOLDER, fil), 'r')
    print 'Writing code for file '+fil
    txt = f.read()
    soup = BeautifulSoup(txt, 'html.parser')
    tags = soup.find_all('pre')
    code_file_path = os.path.join(CODE_OUTPUT_FOLDER, str(fil))
    
    # Create a new empty file
    open(code_file_path, 'w').close()
    for i in tags:
        text =  i.get_text()
        #if text.strip()[0].isalpha():
        with open(code_file_path, 'a') as code_text:
            code_text.write(text)
    f.close()
```

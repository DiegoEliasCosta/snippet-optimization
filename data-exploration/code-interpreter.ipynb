```python
import pandas as pd
import os

# Constants
QUESTION_TYPE = 1
ANSWER_TYPE = 2

file = os.path.join('..', 'data', 'pandas-code-dataset.csv')

dataset = pd.read_csv(file, encoding='ISO-8859-1', error_bad_lines=False, sep=";")
print('Amount of entries in the dataset = %d' % len(dataset))
dataset
```

```python
import ast

questions = dataset[dataset.PostTypeId == QUESTION_TYPE]

def containNaturallyParseable(df):
    strcodes = df.Code
    codes = ast.literal_eval(strcodes)
    for code in codes:
        # If any code is parseable = TRUE
        if parseCode(code):
            return True
    return False

def reportNaturallyParseable(df):
    strcodes = df.Code
    codes = ast.literal_eval(strcodes)
    parseableCount = 0
    for code in codes:
        if parseCode(code):
            parseableCount += 1
    return '%d / %d' % (parseableCount, len(codes))


def parseCode(code):
    try:
        tree = ast.parse(code)
        return True
    except SyntaxError:
        return False
    
   
answers = dataset[dataset.PostTypeId == ANSWER_TYPE]
answers['Contains NP'] = answers.apply(containNaturallyParseable, axis=1) 
answers['NP Report'] = answers.apply(reportNaturallyParseable, axis=1) 

pd.set_option('display.max_colwidth', -1)
answers[['Code', 'Contains NP', 'NP Report']]

    
```

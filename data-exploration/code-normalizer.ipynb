```python
import pandas as pd
import os
import ast
import pyflakes

file = os.path.join('..', 'data', 'pandas-processedcode-dataset.csv')

dataset = pd.read_csv(file, encoding='ISO-8859-1', error_bad_lines=False, sep=";")
print('Amount of entries in the dataset = %d' % len(dataset))
dataset
```

```python
import re

def addPandasImport(post, col):
    strcodes = post[col]
    codes = ast.literal_eval(strcodes)
    processed = []
    normalized = False
    for code in codes:
        if "pd." in code and "import pandas as pd" not in code:
            code = "import pandas as pd\n" + code
            normalized = True
            
        if " DataFrame" in code and "import DataFrame" not in code:
            code = "from pandas import DataFrame\n" + code
            normalized = True
            
        processed.append(code)
        
    post['PandasImportCode'] = processed
    post['PandasImportNormalized'] = normalized
    # Test 
    return post


def addDataFrameDefinition(post, col):
    strcodes = post[col]
    codes = ast.literal_eval(strcodes)
    processed = []
    normalized = False
    for code in codes:
        
        if "df" in code and "df =" not in code:
            code = "df = pd.DataFrame()\n" + code
            normalized = True
            
        if "df1" in code and "df1 =" not in code:
            code = "df1 = pd.DataFrame()\n" + code
            normalized = True
            
        if "df2" in code and "df2 =" not in code:
            code = "df2 = pd.DataFrame()\n" + code
            normalized = True
            
        if "data" in code and "data =" not in code:
            code = "data = pd.DataFrame()\n" + code
            normalized = True
            
        if "data1" in code and "data1 =" not in code:
            code = "data1 = pd.DataFrame()\n" + code
            normalized = True
            
        if "data2" in code and "data2 =" not in code:
            code = "data2 = pd.DataFrame()\n" + code
            normalized = True 
            
        code = "import pandas as pd\n" + code
        processed.append(code)
        
    post['DataFrameCode'] = processed
    post['DataFrameNormalized'] = normalized
    return post



normalized_dataset = dataset.apply(addPandasImport, axis=1, args=('ProcessedCode',))
normalized_dataset = normalized_dataset.apply(addDataFrameDefinition, axis=1, args=('ProcessedCode',))

normalized_dataset.to_csv(os.path.join('..', 'data', 'pandas-normalizedcode-dataset.csv'), encoding='ISO-8859-1', sep=';')
normalized_dataset
```

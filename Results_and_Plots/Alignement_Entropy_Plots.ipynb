```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Word Entropy

### Baseline

```python
baseline_file = 'alignment_entropy/WordEntropy/baseline.csv'
baseline = pd.read_csv(baseline_file, header=None)
print('number of entropies:', len(baseline))
baseline[1] = np.log(1 + baseline[0])
print('median: ', baseline.median(axis=0))
baseline[0].plot(kind='kde')
```

### H1

```python
h1_file = 'alignment_entropy/WordEntropy/h1.csv'
h1 = pd.read_csv(h1_file, header=None)
print('number of entropies:', len(h1))
h1[1] = np.log(1 + h1[0])
print('median: ', h1.median(axis=0))
h1[0].plot(kind='kde')
```

### h2

```python
h2_file = 'alignment_entropy/WordEntropy/h2.csv'
h2 = pd.read_csv(h2_file, header=None)
print('number of entropies:', len(h2))
h2[1] = np.log(1 + h2[0])
print('median: ', h2.median(axis=0))
h2[0].plot(kind='kde')
```

### h1h2

```python
h1h2_file = 'alignment_entropy/WordEntropy/h1h2.csv'
h1h2 = pd.read_csv(h1h2_file, header=None)
print('number of entropies:', len(h1h2))
h1h2[1] = np.log(1 + h1h2[0])
print('median: ', h1h2.median(axis=0))
h1h2[0].plot(kind='kde')
```

### m1

```python
m1_file = 'alignment_entropy/WordEntropy/m1.csv'
m1 = pd.read_csv(m1_file, header=None)
print('number of entropies:', len(m1))
m1[1] = np.log(1 + m1[0])
print('median: ', m1.median(axis=0))
m1[0].plot(kind='kde')
```

## Consolidate result

```python
consolidated_df = pd.DataFrame(columns=['baseline', 'h1', 'h2', 'h1h2', 'm1'])
```

```python
consolidated_df['baseline'] = baseline[0]
consolidated_df['h1'] = h1[0]
consolidated_df['h2'] = h2[0]
consolidated_df['h1h2'] = h1h2[0]
consolidated_df['m1'] = m1[0]
```

```python
consolidated_df.median(axis=0)
```

```python
consolidated_df.describe()
```

```python
fig, axes = plt.subplots(figsize=(8, 5))
entropy_box_plot = sns.boxplot(data = consolidated_df, whis=1, fliersize=0.5, color='white', orient='h') 
```

```python
entropy_box_plot_fig= entropy_box_plot.get_figure()
entropy_box_plot_fig.savefig('berkeley_alignment_entropy_boxplot.pdf')
```
